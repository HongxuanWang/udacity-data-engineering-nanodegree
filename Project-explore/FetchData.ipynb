```python
from craigslist import CraigslistJobs
```

```python
CraigslistJobs.show_filters()
```

```python
import json
import time
timestr = time.strftime("%Y%m%d-%H%M%S")
```

```python
cl_j = CraigslistJobs( filters = {'posted_today':True})

with open(f'jobsraw_{timestr}.jsonl', 'w') as outfile:
    for result in cl_j.get_results():
        json.dump(result, outfile)
        outfile.write('\n')
```
